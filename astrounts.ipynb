```python
df = df.drop(["Profile.Name","Profile.Selection.Group","Mission.Name","Mission.Vechicles.Ascent","Mission.Vechicles.Orbit","Mission.Vechicles.Decent"],axis=1)
```

```python
df['Age'] = df['Mission.Year'] - df['Profile.Birth Year']
df.head()
```

```python
one_hot_encoded_data = pd.get_dummies(df, columns = ['Profile.Gender', 'Profile.Military','Mission.Role','Profile.Nationality'])
one_hot_encoded_data
```

```python
x = one_hot_encoded_data.drop("Age",axis=1)
y = one_hot_encoded_data["Age"]
```

```python
x.head()
```

```python
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
X_scaler=ss.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X_scaler,y, test_size=0.2, random_state=7)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr_model=lr.fit(X_train,y_train)
```

```python
y_pred=lr_model.predict(X_test)
y_pred
```

```python
print(y_pred[0:10])
print(y_test[0:10])
```

```python
lr_model.score(X_test,y_test)
```
